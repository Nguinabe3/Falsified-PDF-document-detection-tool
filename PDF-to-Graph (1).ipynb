# PDF to Graph using pdf reader

```python
from pdfreader import PDFDocument
import networkx as nx
import matplotlib.pyplot as plt

# file_name = "informatique.pdf"

def pdf_to_graph (file_name):
    
    fd = open(file_name, "rb")
    document = PDFDocument(fd)
    

    
    ROOT = "Root"
    
    try:
        PDF_VERSION = document.header.version # Version du pdf
    except:
        pass
    finally:
        pass
    
    try:
        ROOT_TYPE = document.root.Type # Type du root : Catalog
    except:
        pass
    finally:
        pass
    
    try:
        ROOT_LANG = str(document.root.Lang) #Langue utilisee dans le document
    except:
        pass
    finally:
        pass
    
    try:
        ROOT_STRUCT_TREE_ROOT = str(document.root.StructTreeRoot) #
    except:
        pass
    finally:
        pass
    
    try:
        ROOT_METADATA = str(document.root.Metadata.Subtype) #Metadata du document
    except:
        pass
    finally:
        pass
    
    try:
        ROOT_VIEWER = str(document.root.ViewerPreferences) 
    except:
        pass
    finally:
        pass
    
    PAGES = "Pages"

    try:
        PAGES_TYPE = document.root.Pages.Type # Type: Pages
    except:
        pass
    finally:
        pass
    
    try:
        NUM_PAGES = str(document.root.Pages.Count)  # Nombre de pages
    except:
        pass
    finally:
        pass

    PAGES_KIDS  = "Kids"

    try:
        KIDS_TYPE = document.root.Pages.Kids[0].Type
    except:
        pass
    finally:
        pass



    #  Group bloc

    GROUP = "Group"
    
    try:
        GROUP_TYPE = document.root.Pages.Kids[0].Group.Type
    except:
        pass
    finally:
        pass

    try:
        GROUP_CHILD_1 = document.root.Pages.Kids[0].Group.S
    except:
        pass
    finally:
        pass

    try:
        GROUP_CHILD_2 = document.root.Pages.Kids[0].Group.CS
    except:
        pass
    finally:
        pass

    # --- End Group bloc


    # MediaBox bloc

    MEDIABOX = "MediaBox"
    try:
        MEDIABOX_CHILD_1 = document.root.Pages.Kids[0].MediaBox[0]
    except:
        pass
    finally:
        pass
    
    try:
        MEDIABOX_CHILD_2 = document.root.Pages.Kids[0].MediaBox[1]
    except:
        pass
    finally:
        pass

    try:
        MEDIABOX_CHILD_3 = document.root.Pages.Kids[0].MediaBox[2]
    except:
        pass
    finally:
        pass
    
    try:
        MEDIABOX_CHILD_4 = document.root.Pages.Kids[0].MediaBox[3]
    except:
        pass
    finally:
        pass


    # end MediaBox bloc


    # Contents bloc

    CONTENTS = "Contents"


    # end Contents 


    # StructParents bloc

    STRUCTPARENTS = "StructParents"

    try:
        STRUCTPARENTS_CHILD = document.root.Pages.Kids[0].StructParents
    except:
        pass
    finally:
        pass


    # end StructParents bloc


    # Tabs bloc

    TABS = "Tabs"

    try:
        TABS_CHILD = document.root.Pages.Kids[0].Tabs
    except:
        pass
    finally:
        pass

    # end Tabs bloc


    # Resources Bloc

    RESOURCES = "Ressources"


    # --- Resources - Font bloc

    RESOURCES_FONT  = "Font"

    FONT_1 = "F1"


    # --- Font_1

    try:
        BASE_FONT = document.root.Pages.Kids[0].Resources.Font.F1.BaseFont
    except:
        pass
    finally:
        pass
    
    try:
        ENCONDING = document.root.Pages.Kids[0].Resources.Font.F1.Encoding
    except:
        pass
    finally:
        pass
    
    try:
        FIRST_CHAR = str(document.root.Pages.Kids[0].Resources.Font.F1.FirstChar)
    except:
        pass
    finally:
        pass
    
    try:
        LAST_CHAR = str(document.root.Pages.Kids[0].Resources.Font.F1.LastChar)
    except:
        pass
    finally:
        pass
    
    try:
        WIDTHS = str(document.root.Pages.Kids[0].Resources.Font.F1.Widths)
    except:
        pass
    finally:
        pass

    # Font Descriptor
    
    try:
        FONT_NAME = document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.FontName
    except:
        pass
    finally:
        pass
    
    try:
        FLAGS = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.Flags)
    except:
        pass
    finally:
        pass
        
    try:
        ITALIC_ANGLE = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.ItalicAngle)
    except:
        pass
    finally:
        pass
        
    try:
        ASCENT = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.Ascent)
    except:
         pass
    finally:
         pass
    
    try:
        DESCENT = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.Descent)
    except:
        pass
    finally:
        pass
    
        
    try:
        CAP_HEIGHT = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.CapHeight)
    except:
         pass
    finally:
         pass
    
    try:
        AVGWIDTH = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.AvgWidth)
    except:
        pass
    finally:
        pass
        
    try: 
        MAX_WIDTH = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.MaxWidth)
    except:
        pass
    finally:
        pass
        
    try:
        FONT_WEIGHT = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.FontWeight)
    except:
        pass
    finally:
        pass
    
    try:
        XHEIGHT = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.XHeight)
    except:
        pass
    finally:
        pass    
    try:
        LEADING =  str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.Leading)
    except:
        pass
    finally:
        pass
    
    try:
        STEMV = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.StemV)
    except:
        pass
    finally:
        pass
    
    try:
        FONT_B_BOX = str(document.root.Pages.Kids[0].Resources.Font.F1.FontDescriptor.FontBBox)
    except:
        pass
    finally:
        pass

    # end Font Descriptor


    # --- end Font_1
    

    # --- Resources - end Font bloc


    # --- Resources - ExtGstate bloc

    EXTGSTATE = "ExtGstate"
    
    try:
        EXTGSTATE_CHILD_1 ="GS7 " + str(document.root.Pages.Kids[0].Resources.ExtGState.GS7)
    except:
        pass
    finally:
        pass


    try:
        EXTGSTATE_CHILD_2 ="GS8 " + str(document.root.Pages.Kids[0].Resources.ExtGState.GS8)
    except:
        pass
    finally:
        pass


    # --- Resources - end ExtGstate bloc


    # --- Resources - ProcSet bloc

    PROCSET = "ProcSet"

    try:
        PROCSET_CHILD_1 = document.root.Pages.Kids[0].Resources.ProcSet[0]
    except:
        pass
    finally:
        pass

    try:
        PROCSET_CHILD_2 = document.root.Pages.Kids[0].Resources.ProcSet[1]
    except:
        pass
    finally:
        pass

    try:
        PROCSET_CHILD_3 = document.root.Pages.Kids[0].Resources.ProcSet[2]
    except:
        pass
    finally:
        pass

    try:
        PROCSET_CHILD_4 = document.root.Pages.Kids[0].Resources.ProcSet[3]
    except:
        pass
    finally:
        pass

    try:
        PROCSET_CHILD_5 = document.root.Pages.Kids[0].Resources.ProcSet[4]
    except:
        pass
    finally:
        pass

    # --- Resources - end ProcSet bloc



    # End Resources bloc


    # Plotting the graph

    G = nx.Graph()

    graph = [
    (ROOT_TYPE, PDF_VERSION), 
    (ROOT_TYPE, ROOT_LANG),
    (ROOT_TYPE, ROOT_METADATA),
    (ROOT_TYPE, PAGES_TYPE),
    (PAGES_TYPE, NUM_PAGES),
    (PAGES_TYPE, KIDS_TYPE),
    (PAGES_KIDS, RESOURCES),
    (RESOURCES, RESOURCES_FONT),
    (RESOURCES_FONT, FONT_1),
    (FONT_1, BASE_FONT),(FONT_1,ENCONDING),(FONT_1,FIRST_CHAR),(FONT_1,LAST_CHAR),(FONT_1,WIDTHS),
    (FONT_1, "Font Descriptor"),
    ("Font Descriptor", FONT_NAME),
    ("Font Descriptor",FLAGS),
    ("Font Descriptor",ITALIC_ANGLE),
    ("Font Descriptor", ASCENT),
    ("Font Descriptor", DESCENT),
    ("Font Descriptor", CAP_HEIGHT),
    ("Font Descriptor", AVGWIDTH),
    ("Font Descriptor", MAX_WIDTH),
    ("Font Descriptor", FONT_WEIGHT),
    ("Font Descriptor", XHEIGHT),
    ("Font Descriptor", LEADING),
    ("Font Descriptor", STEMV),
    ("Font Descriptor", FONT_B_BOX),
    (RESOURCES, EXTGSTATE),
    (EXTGSTATE, EXTGSTATE_CHILD_1 ), (EXTGSTATE, EXTGSTATE_CHILD_2) , (RESOURCES, PROCSET),
    (PROCSET, PROCSET_CHILD_1),(PROCSET, PROCSET_CHILD_2),(PROCSET, PROCSET_CHILD_3),(PROCSET, PROCSET_CHILD_4),
    (PAGES_KIDS, MEDIABOX),(MEDIABOX, MEDIABOX_CHILD_1),(MEDIABOX, MEDIABOX_CHILD_2),(MEDIABOX, MEDIABOX_CHILD_3),
    (MEDIABOX, MEDIABOX_CHILD_4), (PAGES_KIDS, STRUCTPARENTS),(STRUCTPARENTS, STRUCTPARENTS_CHILD),(PAGES_KIDS, GROUP ),
    (GROUP, GROUP_TYPE ), (GROUP, GROUP_CHILD_1 ), (GROUP, GROUP_CHILD_2 ), 
    (PAGES_KIDS, CONTENTS)
    ]

    G.add_edges_from(graph)

    options = {
        'node_color': 'green',
        'node_size': 500,
        'width': 1,
    }
    plt.figure(figsize=(15,10))

    nx.draw(G, with_labels = False, **options )
    plt.show()
    return G
```

## CONTENT_TO_GRAPH

```python
from pdfreader import PDFDocument, SimplePDFViewer
import networkx as nx
import matplotlib.pyplot as plt

def pdf_content_to_graph (file_name):
    
    
    fd = open(file_name, "rb")
    document = PDFDocument(fd)
    NUM_PAGES = document.root.Pages.Count
    fd = open(file_name, "rb")
    document = PDFDocument(fd)
    viewer = SimplePDFViewer(fd)
    i=1
    while i<= NUM_PAGES:
        
        viewer.navigate(i)
        viewer.render()
        result = viewer.canvas.strings
        x = ''.join(result)
        text = x.split()
        i+=1

    G = nx.Graph()
    tab = []
    for i in range(len(text) - 1):
        tab.append((text[i],text[i + 1]))

    G.add_edges_from(tab)
    options = {
            'node_color': 'green',
            'node_size': 500,
            'width': 1,
        }
    plt.figure(figsize=(15,10))

    nx.draw(G, with_labels = True, **options )
    plt.show()
    return G
#print(viewer.canvas.strings)
```

```python
G1 = pdf_content_to_graph ("ajout.pdf")
```

```python
G2 = pdf_content_to_graph ("suppression.pdf")
```

```python
nx.is_isomorphic(G1, G2)
```

```python
def compare_content(tab1, tab2):
    temp = []
    for text1, text2 in zip(tab1, tab2):
        if(text1 == text2):
            temp.append(True)
        else:
            temp.append(False)
    if False in temp:
        return False
    else:
        return True
            
            
```

```python
def compare_pdf(file1, file2):
    G1 = pdf_content_to_graph (file1)
    G2 = pdf_content_to_graph (file2)
    if (nx.is_isomorphic(G1, G2) == False):
        return False
    else:
        fd1 = open(file1, "rb")
        document1 = PDFDocument(fd1)
        NUM_PAGES = document1.root.Pages.Count
        
        fd2 = open(file2, "rb")
        document2 = PDFDocument(fd2)
        NUM_PAGES = document2.root.Pages.Count
        
        viewer1 = SimplePDFViewer(fd1)
        i=1
        while i<= NUM_PAGES:
            
            viewer1.navigate(i)
            viewer1.render()
            result1 = viewer1.canvas.strings
            x1 = ''.join(result1)
            text1 = x1.split()
            i+=1
        
        #Second element
        viewer2 = SimplePDFViewer(fd2)
        i=1
        while i<= NUM_PAGES:
            
            viewer2.navigate(i)
            viewer2.render()
            result2 = viewer2.canvas.strings
            x2 = ''.join(result2)
            text2 = x2.split()
            i+=1
        
        # Comparing content function
        return compare_content(text1, text2)
        
        
```

```python
file1 = 'Lettre de Motivation.pdf'
file2 = 'point.pdf'
compare_pdf(file1, file2 )
```
